```python
import pandas as pd
import numpy as np

def evaluate_quadratic(Q_matrix, c_vector, xO):
    Q = np.array(eval(Q_matrix))
    c = np.array(eval(c_vector))
    return Q @ xO + c

def gradient_rosenbrock(x):
    x1, x2 = x[0][0], x[1][0]
    return np.array([[400 * x1**3 - 400 * x1 * x2 + 2 * x1 - 2], [200 * x2 - 200 * x1**2]])

def gradient_descent(function, xO, tolerance, max_iterations, step_size_type, step_size_value):
    xO = xO.astype(float)
    iterations = []
    x_values = []
    p_values = []
    error_values = []
    
    for k in range(max_iterations):
        gradient = function(xO)
        direction = -gradient
        alpha = step_size_value  # Utiliza el valor proporcionado por el usuario
        xO += alpha * direction
        eps = np.linalg.norm(gradient, ord=2)
        
        iterations.append(k)
        x_values.append(xO.round(2).tolist())
        p_values.append(direction.round(2).tolist())
        error_values.append(np.format_float_scientific(eps, precision=4))
        
        if eps <= tolerance:
            break
    
    return pd.DataFrame({'Iter': iterations, 'Xk': x_values, 'Pk': p_values, 'Error': error_values})
```

```python

import numpy as np

# Definir la función cuadrática Q y el vector c
Q_matrix = "[[4, 2], [2, 2]]"
c_vector = "[1, 1]"
initial_x = np.array([0.0, 0.0])  # Punto de inicio como arreglo NumPy
tolerance = 1e-6
max_iterations = 1000
step_size_type = 0  # Tipo de tamaño de paso
step_size_value = 0.01

# Define una función lambda que llama a evaluate_quadratic con los argumentos adecuados
result = gradient_descent(lambda x: evaluate_quadratic(Q_matrix, c_vector, x), initial_x, tolerance, max_iterations, step_size_type, step_size_value)

```
